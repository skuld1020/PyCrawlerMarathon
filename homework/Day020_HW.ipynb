### 根據範例的教學，請找到Dcard API，並利用 Requests 呼叫「取得所有看板」，完成資料抓取。

先看感情板：

https://www.dcard.tw/service/api/v2/forums/relationship/posts?popular=true&limit=30&before=234075824  <br>
其中感情板的板名為relationship

再看心情板

https://www.dcard.tw/service/api/v2/forums/mood/posts?popular=true&limit=30&before=234072380  <br>
其中心情板板名為mood

合理推測看板的api為https://www.dcard.tw/service/api/v2/forums

```python
import requests
import json

url_p = 'https://www.dcard.tw/service/api/v2/forums/'
headers = {'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Mobile Safari/537.36'}

r = requests.get(url_p,headers = headers)
r = json.loads(r.text)
r
```

```python
for i in r:
    print('版名:{}'.format(i['name']))
    print('敘述:{}'.format(i['description']))
    print('')
```
