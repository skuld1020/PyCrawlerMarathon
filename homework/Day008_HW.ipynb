### 利用 Request + BeatifulSoup 爬取下列兩個網站內容並解析：

* 1. Dcared 網址： https://www.dcard.tw/f

* 2. 知乎： https://www.zhihu.com/explore

並且回答下面問題：
1. Request 取回之後該怎麼取出資料，資料型態是什麼？

```python
import requests
from bs4 import BeautifulSoup

url01 = 'https://www.dcard.tw/f'
d01 = requests.get(url01)
print(type(d01.text))

url02 = 'https://www.zhihu.com/explore'
d02 = requests.get(url02)
print(type(d02.content))
```

2. 為什麼要使用 BeatifulSoup 處理？

* 1.可以漂亮的整理出資料內容，若有html中有tag沒有打完善也能主動修正。
* 2.資料內容會以文本型式出現，可以用tag來抓取想要的內文或是屬性。

3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
from bs4 import BeautifulSoup

headers = {'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N)\
                          AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Mobile Safari/537.36'}
url = 'https://www.zhihu.com/explore'
response = requests.get(url,headers = headers)
html = response.content
soup = BeautifulSoup(html,'html.parser')
soup
```

### .text和.content的差別
* .text的類型是str，默認編碼用"ISO-8859-1"
* .content的類型是byte。

所以有時用.text返回會有亂碼，可以使用response.encoding = 'utf-8'修正。
