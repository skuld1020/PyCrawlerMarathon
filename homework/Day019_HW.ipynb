### 動態網頁爬蟲 - 使用 Selenium
了解 Selenium 用於動態網頁爬蟲的原理<br>
能夠使用 Selenium 撰寫動態網頁爬蟲<br>
使用 Requests 爬蟲<br>
目標網址：https://hahow.in/courses

```python
import requests
from bs4 import BeautifulSoup

r = requests.get('https://hahow.in/courses')
soup = BeautifulSoup(r.text,'lxml')

#截取soup的正文(body.text)，發現只有javascript的語法，沒有文章
soup.body.text
```

### 動態模擬瀏覽器行為

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select 
from bs4 import BeautifulSoup
```

```python
browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get('https://hahow.in/courses')
```

```python
#取得資料
html_source = browser.page_source
html_source
```

```python
soup = BeautifulSoup(html_source,'html.parser')
soup
#利用selenium就能讀取資料
```

```python
soup.find('h4',class_= 'title').text
browser.close()
```
