### 找一個需要登入的網站試試看，並說明思考流程
（如果不知道要用哪個網站的話，可以試試看 https://github.com/new 網址，未登入時會被導向登入頁）<br>
以下先以ptt八卦板為例

```python
#未加session 或 cookie的情況下

import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
res = requests.get(url)

soup = BeautifulSoup(res.text,'html.parser')
soup
```

```python
import requests
from bs4 import BeautifulSoup

#利用session
rs = requests.session()
payload = {'from':'/bbs/Gossiping/index.html',
          'yes':'yes'}

#在requests做请求的时候，为了避免ssl认证，可以放入參數verify=False(但是日誌會報warning，不影響結果)
#先送請求模擬登入(確認滿18歲)
#為了將報錯訊息去掉，多加requests.packages.urllib3.disable_warnings()即可
requests.packages.urllib3.disable_warnings()
res = rs.post('https://www.ptt.cc/ask/over18',verify=False,data = payload)

#再次發送請求網頁資訊
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
res = rs.get(url)

soup = BeautifulSoup(res.text,'html.parser')
soup
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
#直接加入cookies
cookies = {'over18':'1'}
res = requests.get(url,cookies=cookies)

soup = BeautifulSoup(res.text,'html.parser')
soup
```
