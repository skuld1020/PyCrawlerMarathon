```python
import requests
import re
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
html = requests.get(url)
soup = BeautifulSoup(html.text,'html.parser')

a = soup.find(class_="part_list_2").find_all('h3')
pattern = re.compile('tag\sc\w+')

#若依照分類，可以用字典的形式儲存。

news=dict()

for i in a:
    date = i.find(class_='date').text
    title = i.find('a').text
    tag = i.find(class_=pattern).text
    news.setdefault(tag,[])
    news[tag].append({date:title})
print(news)   
```

```python
#上面跑出來的結果有'\u3000'為網頁原使碼，可用replace替換
import requests
import re
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
html = requests.get(url)
soup = BeautifulSoup(html.text,'html.parser')

a = soup.find(class_="part_list_2").find_all('h3')
pattern = re.compile('tag\sc\w+')

#若依照分類，可以用字典的形式儲存。

news=dict()

for i in a:
    date = i.find(class_='date').text
    title = i.find('a').text
    #用''取代\u3000
    title = title.replace('\u3000','')
    tag = i.find(class_=pattern).text
    news.setdefault(tag,[])
    news[tag].append({date:title})
print(news)   
```

### dict.setdefault()
setdefault()方法類似於get()方法，但會設置字典[鍵]=默認情況下，如果鍵不是已經在字典中。
dict.setdefault(key, default=None)

* key -- 這是要被搜索的鍵
* default -- 這是冇有找到鍵的情況下返回的值。

### replace()方法语法：

str.replace(old, new[, max])

* old -- 将被替换的子字符串。
* new -- 新字符串，用于替换old子字符串。
* max -- 可选字符串, 替换不超过 max 次

```python
import requests
import re
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
html = requests.get(url)
soup = BeautifulSoup(html.text,'html.parser')

a = soup.find(class_="part_list_2").find_all('h3')
pattern = re.compile('tag\sc\w+')

#新聞類型最多的是
total = dict()
for i in a:
    total[i.find(class_=pattern).text] = total.get(i.find(class_=pattern).text,0)+1
print(total)
M = max(total.values())
m = min(total.values())
for k,v in total.items():
    if total[k] == M:
        C = k
    if total[k] == m:
        c = k

print('最多類型的是{}，一共有{}則；最少類型的是{}，一共有{}則'.format(C,M,c,m))
```
