### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
#r.text

import requests

r = requests.get('https://api.github.com/events')
r.text
```

```python
#json.loads(r.text)
import requests
import json

r = requests.get('https://api.github.com/events')
json.loads(r.text)
```

### 自行尋找一個合適的 API 接口做練習，並且查看其回傳內容：
* https://cat-fact.herokuapp.com/facts (來源：https://alexwohlbruck.github.io/cat-facts/)
* http://odata.wra.gov.tw/v4/RealtimeWaterLevel (來源：https://data.gov.tw/dataset/25768)

```python
import requests
import json

s = requests.get('https://cat-fact.herokuapp.com/facts')
json.loads(s.text)
```

```python
import requests
import json

s = requests.get('http://odata.wra.gov.tw/v4/RealtimeWaterLevel')
json.loads(s.text)
```
