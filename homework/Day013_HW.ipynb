### 根據範例 ，完成以下問題：
* 印出最新文章的「作者」「標題」「時間」
* 印出第一頁所有文章的「作者」「標題」「時間」
* 試著爬爬看其他版的文章

目標網址：https://www.ptt.cc/bbs/NBA/index.html

```python
import requests
from bs4 import BeautifulSoup

url1 = 'https://www.ptt.cc/bbs/NBA/index.html'
html = requests.get(url1)
soup = BeautifulSoup(html.text,'html.parser')

#尋找最新文章的網址
tag = soup.find(class_="title")
id = tag.find('a').get('href')
url = 'https://www.ptt.cc'+id

#印出最新文章的「作者」「標題」「時間」
data = requests.get(url)
soup = BeautifulSoup(data.text,'html.parser')

#作者、標題和日期都有<span class="article-meta-value">這個tag
d = soup.select('span.article-meta-value')

title = d[2].text
author = d[0].text
date = d[3].text
name = d[1].text

print('標題：',title)
print('看板：',name)
print('作者：',author,'   日期：',date)
print(url)
```

```python
#印出第一頁所有文章的「作者」「標題」「時間」
import requests
from bs4 import BeautifulSoup

url1 = 'https://www.ptt.cc/bbs/NBA/index.html'
html = requests.get(url1)
soup = BeautifulSoup(html.text,'html.parser')

data = soup.find_all(class_='r-ent')

for i in data:
    title = i.find(class_='title').text
    author = i.find(class_='author').text
    date = i.find(class_='date').text
    print(title,author,date)
```

### 爬取八卦版3頁文章標題 & 網址
目標網址：https://www.ptt.cc/bbs/Gossiping/index.html


```python
import requests
from bs4 import BeautifulSoup

#定義能抓取八卦版一頁所有連結與標題的函數

def page_title_links(url):   
    res = requests.get(url,cookies={'over18':'1'})
    soup = BeautifulSoup(res.text,'html.parser')

    #選出所有div class='title'的tag
    data = soup.select('div.title')
    for i in data:
        item =i.select_one('a')
        #怕有文章被刪，要先確認item是否存在再進行取網址行為
        if item:
            title = i.text
            link = 'https://www.ptt.cc' + item.get('href')
            print(title,link)


url='https://www.ptt.cc/bbs/Gossiping/index.html'

for page in range(0,3):
    print(url)
    print('---------')
    page_title_links(url)
    res = requests.get(url,cookies={'over18':'1'})
    soup = BeautifulSoup(res.text,'html.parser')
    #八卦板的頁面選擇放在class='btn'裡
    btn = soup.select('.btn')
    next_page = 'https://www.ptt.cc' + btn[3].get('href')
    url = next_page
    
```

### soup.select
select的功能跟find和find_all 一樣用來選取特定的標籤，
它的選取規則依賴於css，我們把它叫做css選擇器，如果之前有接觸過jquery，可以發現select的選取規則和jquery有點像。

* print soup.select('p')  →  返回所有含有'p'的tag
* soup.select('.title') → 類名(class)前加點(class='title')
* print soup.select('#link2') → id名前加#(id='link2')
* print soup.select('[href="http://example.com/lacie"]')  →  選擇href 為http://example.com/lacie　的tag

# 組合查詢
組合查詢可以分為兩種，一種是在一個tag中進行兩個條件的查詢，一種是樹狀的查詢一層一層之間的查詢。

* print soup.select('a#link2')  →  選擇標籤名為a，id為link2的tag。
輸出的結果如下：
* [<a class="sister" href="http://example.com/lacie" id="link2">Lacie</a>]

從body開始，在body裡面查詢 所有的p，在所有的p中查詢 標籤名為a，
id 為link2的tag，這樣像樹狀一層一層的查詢，在分析html結構是是非常常見的。層和層之間用空格分開。

print soup.select('body p a#link2')

* .select返回是一個list,.select_one返回為第一個值。

### 爬取3頁八卦板：文章標題、時間、作者


```python
import requests
from bs4 import BeautifulSoup

def print_all_data(url):
    html = requests.get(url,cookies={'over18':'1'})
    soup = BeautifulSoup(html.text,'html.parser')
    tags = soup.find_all(class_="r-ent")
    for tag in tags:
        href = tag.find('a')
        if href:
            title =tag.find(class_='title').text
            date = tag.find(class_='date').text
            author = tag.find(class_='author').text
            link = 'https://www.ptt.cc' + href.get('href')
            print(title,author,date,link)

url = 'https://www.ptt.cc/bbs/Gossiping/index.html'

for page in range(0,3):
    print(url)
    print_all_data(url)
    html = requests.get(url,cookies={'over18':'1'})
    soup = BeautifulSoup(html.text,'html.parser')
    btn = soup.find_all(class_='btn wide')
    lastpageid = btn[1].get('href')
    lastpagelink = 'https://www.ptt.cc' +lastpageid
    url = lastpagelink
    
```
