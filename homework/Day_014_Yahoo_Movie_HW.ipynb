### 範例示範
* YAHOO電影首頁：https://movies.yahoo.com.tw
* 點選「上映中」的分頁，先確認上映中的電影總共有幾筆，然後逐頁爬取。
* 爬取所有上映中的電影資料：電影名稱、期待度、滿意度、簡介等等。

```python
import requests
from bs4 import BeautifulSoup

#上映中的頁面網址
url ='https://movies.yahoo.com.tw/movie_intheaters.html'

#確認上映中的電影筆數
r = requests.get(url)
soup = BeautifulSoup(r.text,'html.parser')
t = soup.find(class_='release_box').findChild(class_='release_time _c')
print(t.find('p').text)
```

```python
#爬取上映中的電影資料：電影名稱、期待度、滿意度、簡介等等
import requests
from bs4 import BeautifulSoup


def print_alldata(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html.parser')
    movielist = soup.find(class_='release_list').find_all('li')
    for m in movielist:
        name = m.find(class_='release_movie_name')
        #將class_='release_movie_name'底下的<a></a>的內容取出
        name = name.a.string
        name = name.lstrip()
        
        expectation = m.find(class_='leveltext')
        #在class_='leveltext'底下把<span></span>的內容取出
        expectation = expectation.span.string
    
        satisfaction = m.find('dd').find(class_='leveltext')
        #在class_='leveltext'底下把<span></span>裡的data-num數性取出
        satisfaction = satisfaction.span['data-num']
    
        contents = m.find(class_='release_text')
        contents = contents.span.string.lstrip()
    
        print('片名：',name,'期待度：', expectation,'滿意度：',satisfaction)
        print('簡介：')
        print(contents)
        

    
#尋找一共有幾頁
url ='https://movies.yahoo.com.tw/movie_intheaters.html'
res = requests.get(url)
soup = BeautifulSoup(res.text,'html.parser')
pagenumber = soup.find(class_='page_numbox').find_all('a')
n = int(pagenumber[-2].string)

for i in range(1,n+1):
    url = 'https://movies.yahoo.com.tw/movie_intheaters.html?page='+str(i)
    print_alldata(url)
```

### 先搜尋全部的電影代號(ID)資訊

```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://movies.yahoo.com.tw/'
r = requests.get(url)
r.encoding ='utf-8'
soup = BeautifulSoup(r.text,'html.parser')
movieid = soup.find(class_='movie_name',attrs={'name':'movie_id'})
movielist = movieid.find_all('option',attrs = {'data-name':re.compile('.+')})
for i in movielist:
    name = i.get('data-name')
    id = i.get('value')
    print(id,name)
```
